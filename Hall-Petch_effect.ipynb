### The precipitates are assumed as circle and not merging. The domain is L^2 and the total volume fraction of the precipitates are "volume_fraction". R_precipitate is the size of it.

```python
import math

L = 501

volume_fraction = 0.1

R_precipitate = math.sqrt(volume_fraction*(L**2)/math.pi)

X = []
Y1 = []

for j in range(L):
    Z = [] 
    for i in range(L):
        dis = 0.
        dis = math.sqrt((i-0.5*L)**2.+(j-0.5*L)**2.)
        
        if (dis < R_precipitate):
            Z.append(1) 
        else:
            Z.append(0) 
    Y1.append(Z) 

Y = []
Y.append(Y1)

from numpy import array
X = array(Y)

MKSmodel.resize_coeff(X[0].shape)
strain_pred = MKSmodel.predict(X)

draw_microstructure_strain(X[0], strain_pred[0])
```

## Gaussian Distribution of the precipitate size

### The mean size of the precipitates is "R_precipitate". Within the Gaussian domain, it is devided by 11 in this example.

```python
R_precipitate = 30

Gau_R_LB = -5
Gau_R_UB = 6
Gau_no_samples = Gau_R_UB - Gau_R_LB
Gau_R_dis = []

Gau_delta = 1.5
Gau_mu = 0

Gau_temp = []
prob = 0.
prob_tot = 0.

R_temp = 0.
for R_dis in range(Gau_R_LB,Gau_R_UB):
    R_temp = int(R_precipitate + R_dis)

    prob = np.exp(-np.power(R_dis+Gau_mu, 2.) / (2 * np.power(Gau_delta, 2.)) ) \
           / math.sqrt(2*math.pi*np.power(Gau_delta, 2.))

    Gau_temp = (R_temp,prob)
    
    Gau_R_dis.append(Gau_temp)
    
Gau_R_dis = array(Gau_R_dis)
prob_tot = Gau_R_dis.sum(axis=0)

for i in range(1,Gau_no_samples):
    Gau_R_dis[i][1] = Gau_R_dis[i-1][1] + Gau_R_dis[i][1] / prob_tot[1]

UB_Gau = Gau_R_dis[Gau_no_samples-1][1]
```

### Since the size is ramdoml selected, the number of precipitates is not a constant. So is the total volume fraction of them. 

```python
import random

vf_in_domain = 0.
no_particles = 0

R_Gau = []
while (vf_in_domain <= volume_fraction):
    R_ind = random.uniform(0,UB_Gau)
    
    if ((R_ind > 0.) and (R_ind <= Gau_R_dis[0][1])):
        R_Gau.append(Gau_R_dis[0][0])
    else:
        for i in range(1,Gau_no_samples):
            if ((R_ind > Gau_R_dis[i-1][1]) and (R_ind <= Gau_R_dis[i][1])):
                R_Gau.append(Gau_R_dis[i][0])
                break

    vf_in_domain += math.pi*(R_Gau[-1]**2.)/(L**2.)
    no_particles += 1

print "Number of precipitates: ", no_particles
print "Actual volume fraction: ", vf_in_domain
print "Default volume fraction: ", volume_fraction
```

```python
peri_para = L-1
ind_phase1_r = []

p_center = [[] for i in range(no_particles)]

p_x = random.randint(1,L-1)
p_y = random.randint(1,L-1)
p_center[0].append(p_x)
p_center[0].append(p_y)

iR=-int(R_Gau[0])
iL=int(R_Gau[0])

for i in range(iR,iL):
    R_p_x = i + p_x
    if (R_p_x < 0):
        R_p_x += peri_para
    elif (R_p_x > peri_para):
        R_p_x -= peri_para

    for j in range(iR,iL):
        R_p_y = j + p_y
        if (R_p_y < 0):
            R_p_y += peri_para
        elif (R_p_y > peri_para):
            R_p_y -= peri_para

        dis = math.sqrt((R_p_x-p_x)**2.+(R_p_y-p_y)**2.)
        disx1 = math.sqrt((R_p_x+peri_para-p_x)**2.+(R_p_y-p_y)**2.) # periodic BC for x=0 side
        disx2 = math.sqrt((R_p_x-peri_para-p_x)**2.+(R_p_y-p_y)**2.) # periodic BC for x=L side
        disy1 = math.sqrt((R_p_x-p_x)**2.+(R_p_y+peri_para-p_y)**2.) # periodic BC for y=0 side
        disy2 = math.sqrt((R_p_x-p_x)**2.+(R_p_y-peri_para-p_y)**2.) # periodic BC for y=L side
        disc1 = math.sqrt((R_p_x-peri_para-p_x)**2.+(R_p_y-peri_para-p_y)**2.) # corners
        disc2 = math.sqrt((R_p_x-peri_para-p_x)**2.+(R_p_y+peri_para-p_y)**2.) # corners
        disc3 = math.sqrt((R_p_x+peri_para-p_x)**2.+(R_p_y-peri_para-p_y)**2.) # corners
        disc4 = math.sqrt((R_p_x+peri_para-p_x)**2.+(R_p_y+peri_para-p_y)**2.) # corners

        R_test = min(dis,disx1,disx2,disy1,disy2,disc1,disc2,disc3,disc4)

        Rtemp = []
        if (R_test <= R_Gau[0]):
            Rtemp = (R_p_x,R_p_y)
            ind_phase1_r.append(Rtemp)

ind_var = 1
while (ind_var < no_particles):
    p_x = random.randint(1,L-1)
    p_y = random.randint(1,L-1)
    NA = 0
    
    for i in range(0,ind_var):
        dis = math.sqrt((p_x-p_center[i][0])**2.+(p_y-p_center[i][1])**2.)
        
        disx1 = math.sqrt((p_x+peri_para-p_center[i][0])**2.+(p_y-p_center[i][1])**2.) # periodic BC for x=0 side
        disx2 = math.sqrt((p_x-peri_para-p_center[i][0])**2.+(p_y-p_center[i][1])**2.) # periodic BC for x=L side
        disy1 = math.sqrt((p_x-p_center[i][0])**2.+(p_y+peri_para-p_center[i][1])**2.) # periodic BC for y=0 side
        disy2 = math.sqrt((p_x-p_center[i][0])**2.+(p_y-peri_para-p_center[i][1])**2.) # periodic BC for y=L side
        disc1 = math.sqrt((p_x-peri_para-p_center[i][0])**2.+(p_y-peri_para-p_center[i][1])**2.) # corners
        disc2 = math.sqrt((p_x-peri_para-p_center[i][0])**2.+(p_y+peri_para-p_center[i][1])**2.) # corners
        disc3 = math.sqrt((p_x+peri_para-p_center[i][0])**2.+(p_y-peri_para-p_center[i][1])**2.) # corners
        disc4 = math.sqrt((p_x+peri_para-p_center[i][0])**2.+(p_y+peri_para-p_center[i][1])**2.) # corners
       
        R_test = min(dis,disx1,disx2,disy1,disy2,disc1,disc2,disc3,disc4)

        if (R_test < (R_Gau[ind_var] + R_Gau[i]) ):
            NA = 1
            break
            
    if ( NA == 0 ):
        p_center[ind_var].append(p_x)
        p_center[ind_var].append(p_y)
        
        iR=-int(R_Gau[ind_var])
        iL=int(R_Gau[ind_var])
        
        for i in range(iR,iL):
            R_p_x = i + p_x
            if (R_p_x < 0):
                R_p_x += peri_para
            elif (R_p_x > peri_para):
                R_p_x -= peri_para
                
            for j in range(iR,iL):
                R_p_y = j + p_y
                
                if (R_p_y < 0):
                    R_p_y += peri_para
                elif (R_p_y > peri_para):
                    R_p_y -= peri_para
                
                dis = math.sqrt((R_p_x-p_x)**2.+(R_p_y-p_y)**2.)
                disx1 = math.sqrt((R_p_x+peri_para-p_x)**2.+(R_p_y-p_y)**2.) # periodic BC for x=0 side
                disx2 = math.sqrt((R_p_x-peri_para-p_x)**2.+(R_p_y-p_y)**2.) # periodic BC for x=L side
                disy1 = math.sqrt((R_p_x-p_x)**2.+(R_p_y+peri_para-p_y)**2.) # periodic BC for y=0 side
                disy2 = math.sqrt((R_p_x-p_x)**2.+(R_p_y-peri_para-p_y)**2.) # periodic BC for y=L side
                disc1 = math.sqrt((R_p_x-peri_para-p_x)**2.+(R_p_y-peri_para-p_y)**2.) # corners
                disc2 = math.sqrt((R_p_x-peri_para-p_x)**2.+(R_p_y+peri_para-p_y)**2.) # corners
                disc3 = math.sqrt((R_p_x+peri_para-p_x)**2.+(R_p_y-peri_para-p_y)**2.) # corners
                disc4 = math.sqrt((R_p_x+peri_para-p_x)**2.+(R_p_y+peri_para-p_y)**2.) # corners

                R_test = min(dis,disx1,disx2,disy1,disy2,disc1,disc2,disc3,disc4)

                Rtemp = []
                if (R_test <= R_Gau[ind_var]):
                    Rtemp.append(R_p_x)
                    Rtemp.append(R_p_y)
                    ind_phase1_r.append(Rtemp)
                    vf_in_domain += 1
                    
        ind_var += 1


X = []
Y1 = [[0 for x in xrange(L)] for x in xrange(L)]

for i in range(len(ind_phase1_r)):
    Y1[ind_phase1_r[i][0]][ind_phase1_r[i][1]] = 1

Y = []
Y.append(Y1)
X = array(Y)

MKSmodel.resize_coeff(X[0].shape)
strain_pred = MKSmodel.predict(X)

draw_microstructure_strain(X[0], strain_pred[0])
```
